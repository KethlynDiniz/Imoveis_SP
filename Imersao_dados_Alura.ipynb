```python
dados[['Moeda', 'Valor_anuncio', 'Tipo_anuncio']] = dados.Valor.str.split(expand = True) #colocando o expand dentro da nossa base de dados
dados.sample(10)
```

```python
dados[dados['Tipo_anuncio'].isnull()]['Tipo_anuncio'].unique()
```

```python
dados_vendas = dados[dados['Tipo_anuncio'].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas.Valor_anuncio.str.replace('.', '').astype(float)
```

```python
dados_vendas['Valor_anuncio_float'] = dados_vendas.Valor_anuncio.str.replace('.', '').astype(float)
```

```python
dados_vendas.info()
```

```python
dados_vendas.Valor_anuncio_float.plot.hist(bins = 50)
```

```python
dados_vendas.describe()
```

```python
#pd.set_option('display.precision', 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x )  #para transformar da cientifica para 'normal' o float
dados_vendas.describe(include = 'all')
```

```python
dados_vendas[['Rua', 'Bairro', 'Cidade']].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
plt.figure(figsize = (8, 6))
ax = sns.histplot(data = dados_vendas, x = 'Valor_anuncio_float', bins = 1000, kde = True)
ax.set_title('Histograma Valor Imóvel')
plt.xlim((-50, 10000000))
plt.show()
```
